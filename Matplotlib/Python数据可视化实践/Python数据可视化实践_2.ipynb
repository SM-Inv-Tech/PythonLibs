## 使用统计函数绘制简单图形

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from jupyterthemes import jtplot
jtplot.style("gruvboxl")

mpl.rcParams["font.sans-serif"]=["SimHei"]
mpl.rcParams["axes.unicode_minus"]=False
```

### 用于绘制柱状图bar函数

```python
x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [3, 1, 4, 5, 8, 9, 7, 2]

plt.bar(x, y, align="center", color="c", tick_label=["q", "a", "c", "e", "r", "j", "b", "p"], hatch="/")

plt.xlabel("箱子编号")
plt.ylabel("箱子重量(Kg)")

plt.show()
```

### 用于绘制条形图barh()函数

```python
x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [3, 1, 4, 5, 8, 9, 7, 2]

plt.barh(x, y, align="center", color="red", tick_label=["q", "a", "c", "e", "r", "j", "b", "p"], hatch="/")

plt.xlabel("箱子重量(Kg)")
plt.ylabel("箱子编号")

plt.show()
```

### 用于绘制直方图hist()函数

```python
x = np.random.randint(0, 10, 100)
bins = range(0, 11, 1)

plt.hist(x, bins=bins, color="blue", histtype="bar", rwidth=0.9, alpha=0.6)

plt.xlabel("箱子重量(Kg)")
plt.ylabel("销售数量(个)")

plt.show()
```

### 用于绘制饼图pie()函数

```python
kinds = ["简易箱子", "保温箱子", "行李箱子", "密封箱子"]
colors = ["red", "magenta", "blue", "grey"]
sold_nums = [0.05, 0.45, 0.15, 0.35]

plt.pie(sold_nums, labels=kinds, autopct="%3.1f%%", startangle=60, colors=colors)
plt.title("不同类型箱子的销售数量占比")

plt.show()
```

### 用于绘制散点图scatter()函数

```python
x = np.random.randn(100)
y = np.random.randn(100)

plt.scatter(x, y, s=np.power(10*x + 20*y, 2),c=np.random.rand(100), cmap=mpl.cm.RdYlBu, marker="o")

plt.show()
```

用于绘制棉棒图stem()函数

```python
x = np.linspace(0.5, 2*np.pi, 20)
y = np.random.randn(20)

plt.stem(x, y, linefmt="-.", markerfmt="o", basefmt="-")

plt.show()
```

### 用于绘制箱线图boxplot()函数

```python
x = np.random.randn(1000)

plt.boxplot(x)

plt.xticks([1], ["随机数生成器AlphaRM"])
plt.ylabel("随机数值")
plt.title("随机数生成器抗干扰能力的稳定性")

plt.show()
```

### 用于绘制误差棒图errorbar()函数

```python
x = np.linspace(0.1, 0.6, 6)
y = np.exp(x)

plt.errorbar(x, y, fmt="bo:", yerr=0.2, xerr=0.02, c="red")

plt.xlim(0, 0.7)

plt.show()
```
