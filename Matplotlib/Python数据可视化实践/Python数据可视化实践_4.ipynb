## 完善统计图形

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from jupyterthemes import jtplot
jtplot.style("gruvboxl")

mpl.rcParams["font.sans-serif"] = ["SimHei"]
mpl.rcParams["axes.unicode_minus"] = False
```

### 添加图例和标题

```python
x = np.linspace(-2*np.pi, 2*np.pi, 200)
y_1 = np.sin(x)
y_2 = np.cos(x)

plt.plot(x, y_1, linewidth=2, label=r"$\sin(x)$")
plt.plot(x, y_2, linewidth=2, label=r"$\cos(x)$")

plt.xlabel("横坐标名称")
plt.ylabel("纵坐标名称")

plt.legend(loc="lower left")
plt.title("正弦函数和余弦函数的折线图")

plt.show()
```

### 图例精确设置

```python
x = np.arange(0., 2.1, 0.1)
y_1 = np.power(x, 3)
y_2 = np.power(x, 2)
y_3 = np.power(x, 1)

plt.plot(x, y_1, linestyle="-", linewidth=2, label="$x^{3}$")
plt.plot(x, y_2, linestyle="-", linewidth=2, color="red", label="$x^{2}$")
plt.plot(x, y_3, linestyle="-", linewidth=2, color="yellow", label="$x^{1}$")

plt.legend(loc="upper left", bbox_to_anchor=(0.05, 0.95), ncol=3, title="指数函数", shadow=True, fancybox=True)

plt.show()
```
