## 绘制统计图形

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from jupyterthemes import jtplot
jtplot.style("gruvboxl")

mpl.rcParams["font.sans-serif"] = ["SimHei"]
mpl.rcParams["axes.unicode_minus"] = False
```

### 柱状图

```python
x = [1, 2, 3, 4, 5]
y = [6, 10, 4, 5, 1]

# x: 表示柱体标签值；
# y: 柱体高度；
# align: 柱体对齐方式；
# color: 颜色；
# tick_label: 标签刻度值；
# alpha: 柱体的透明度；
plt.bar(x, y, align="center", color="red", tick_label=["A", "B", "C", "D", "E"], alpha=0.6)

plt.xlabel("测试难度")
plt.ylabel("试卷份数")

plt.show()
```

### 条形图barh()函数

```python
x = [1, 2, 3, 4, 5]
y = [6, 10, 4, 5, 1]

plt.barh(x, y, align="center", tick_label=["A", "B", "C", "D", "E"], color="blue", label="试卷总数")

plt.legend(loc="best")

plt.ylabel("测试难度")
plt.xlabel("试卷份数")

plt.show()
```

### 堆积图bar()函数

```python
x = [1, 2, 3, 4, 5]
y_1 = [6, 10, 4, 5, 1]
y_2 = [2, 6, 3, 8, 5]

plt.bar(x, y_1, align="center", color="red", tick_label=["A", "B", "C", "D", "E"], label="班级A")
plt.bar(x, y_2, align="center", color="blue", bottom=y_1, label="班级B")

plt.xlabel("测试难度")
plt.ylabel("试卷份数")

plt.legend()

plt.show()
```

### 堆积条形图barh()函数

```python
x = [1, 2, 3, 4, 5]
y_1 = [6, 10, 4, 5, 1]
y_2 = [2, 6, 3, 8, 5]

plt.barh(x, y_1, align="center", color="red", tick_label=["A", "B", "C", "D", "E"], label="班级A")
plt.barh(x, y_2, align="center", color="blue", left=y_1, label="班级B")

plt.ylabel("测试难度")
plt.xlabel("试卷份数")

plt.legend()

plt.show()
```

### 多数据并列柱状图bar()函数

```python
x = np.arange(5)
y_1 = [6, 10, 4, 5, 1]
y_2 = [2, 6, 3, 8, 5]

bar_width=0.35
tick_label=["A", "B", "C", "D", "E"]

plt.bar(x, y_1, width=bar_width, color="red", align="center", label="班级A", alpha=0.5)
plt.bar(x+bar_width, y_2, width=bar_width, color="blue", align="center", label="班级B", alpha=0.5)

plt.xlabel("测试难度")
plt.ylabel("试卷份数")

plt.xticks(x+bar_width/2, tick_label)

plt.legend(loc="upper right")

plt.show()
```

### 多数据平行条形图barh()函数

```python
x = np.arange(5)
y_1 = [6, 10, 4, 5, 1]
y_2 = [2, 6, 3, 8, 5]

bar_width=0.35
tick_label=["A", "B", "C", "D", "E"]

plt.barh(x, y_1, height=bar_width, color="red", align="center", label="班级A", alpha=0.5)
plt.barh(x+bar_width, y_2, height=bar_width, color="blue", align="center", label="班级B", alpha=0.5)

plt.ylabel("测试难度")
plt.xlabel("试卷份数")

plt.yticks(x+bar_width/2, tick_label)

plt.legend(loc="upper right")

plt.show()
```

### 参数探索hatch参数

```python
x = [1, 2, 3, 4, 5]
y = [6, 10, 4, 5, 1]

plt.bar(x, y, align="center", color="green", tick_label=["A", "B", "C", "D", "E"], hatch="///")

plt.xlabel("测试难度")
plt.ylabel("试卷份数")

plt.show()
```

### 绘制堆积折线图stackplot()函数

```python
x = np.arange(1, 6, 1)
y_1 = [0, 4, 3, 5, 6]
y_2 = [1, 3, 4, 2, 7]
y_3 = [3, 4, 1, 6, 5]

labels = ["BluePlanet", "BrownPlanet", "GreenPlanet"]
colors = ["blue", "brown", "green"]

plt.stackplot(x, y_1, y_2, y_3, labels=labels, colors=colors)

plt.legend(loc="upper left")

plt.show()
```

### 绘制间段条形图broken_bar()函数

```python
plt.broken_barh([(30, 100), (180, 50), (260, 70)], (20, 8), facecolors="#1f78b4")
plt.broken_barh([(60, 90), (190, 20), (230, 30), (280, 60)], (10, 8), facecolors=["#7fc97f", "#beaed4", "#fdc086", "#ffff99"])

plt.xlim(0, 360)
plt.ylim(5, 35)

plt.xlabel("演出时间")

plt.xticks(np.arange(0, 361, 60))
plt.yticks([15, 25], ["歌剧院A", "歌剧院B"])

plt.title("不同地区的歌剧院的演出时间比较")

plt.show()
```

### 绘制阶梯图

```python
x = np.linspace(1, 10, 10)
y = np.sin(x)

plt.step(x, y, color="red", where="pre", linewidth=2)

plt.xlim(0, 11)
plt.ylim(-1.2, 1.2)
plt.xticks(np.arange(1, 11, 1))

plt.show()
```

### 直方图hist()函数

```python
x = np.random.randint(0, 100, 100)
bins = range(0, 101, 10)

plt.hist(x, bins=bins, color="magenta", histtype="bar", rwidth=1.0, edgecolor="black")

plt.xlabel("测试成绩")
plt.ylabel("学生人数")

plt.show()
```

### 堆叠直方图

```python
score_1 = np.random.randint(0, 100, 100)
score_2 = np.random.randint(0, 100, 100)

x = [score_1, score_2]
colors = ["red", "blue"]
labels = ["班级A", "班级B"]

bins = range(0, 101, 10)

plt.hist(x, bins=bins, color=colors, histtype="bar", rwidth=1.0, stacked=True, label=labels)

plt.xlabel("测试成绩(分)")
plt.ylabel("学生人数")

plt.title("不同班级的测试成绩的直方图")
plt.legend(loc="best")

plt.show()
```

### 箱线图boxplot()函数

```python
test_A = np.random.randn(5000)
test_B = np.random.randn(5000)

test_list = [test_A, test_B]
labels = ["随机数生成器A", "随机数生成器B"]
colors = ["blue", "red"]

bplot = plt.boxplot(test_list, whis=1.6, widths=0.35, sym="o", labels=labels, patch_artist=True)
for patch, color in zip(bplot["boxes"], colors):
    patch.set_facecolor(color)

plt.ylabel("随机数值")
plt.title("生成器抗干扰能力的稳定性比较")

plt.show()
```

### 水平方向的箱线图

```python
x = np.random.randn(1000)

plt.boxplot(x, vert=False)

plt.xlabel("随机数值")
plt.yticks([1], ["随机数生成器AlphaRM"], rotation=90)
plt.title("随机数生成器抗干扰能力的稳定性")

plt.show()
```

### 箱线图不显示离散值

```python
x = np.random.randn(1000)

plt.boxplot(x, vert=False, showfliers=False)

plt.xlabel("随机数值")
plt.yticks([1], ["随机数生成器AlphaRM"], rotation=90)
plt.title("随机数生成器抗干扰能力的稳定性")

plt.show()
```

### 误差图errorbar()函数

```python
x = np.linspace(0.1, 0.6, 10)
y = np.exp(x)

error = 0.05 + 0.15 * x

lower_error = error
upper_error = 0.3 * error
error_limit = [lower_error, upper_error]

plt.errorbar(x, y, yerr=error_limit, fmt=":s", ecolor="green", elinewidth=2, markersize=10, markerfacecolor="red", markeredgecolor="black", capthick=2, capsize=4)
plt.xlim(0, 0.7)

plt.show()
```
