## Python数据可视化时间

```python
import numpy as np
import matplotlib.pyplot as plt
from jupyterthemes import jtplot
jtplot.style("gruvboxl")
```

### plot函数画趋势

```python
x = np.linspace(0.05, 10, 1000)
y = np.cos(x)

plt.plot(x, y, color="red", linestyle="-", linewidth=2, label="plot figure")
plt.legend()

plt.show()
```

### scatter函数

```python
x = np.linspace(0.05, 10, 1000)
y = np.random.rand(1000)

plt.scatter(x, y, label="scatter figure")
plt.legend(loc="upper left")

plt.show()
```

### xlim, ylim函数设置坐标轴范围

```python
x = np.linspace(0.5, 10, 1000)
y = np.random.rand(1000)

plt.scatter(x, y, label="scatter figure")
plt.legend(loc="upper right")

plt.xlim(0.05, 10)
plt.ylim(0, 1)

plt.show()
```

### xlabel函数设置

```python
x = np.linspace(0.05, 10, 1000)
y = np.sin(x)

plt.plot(x, y, linestyle="-", linewidth=1, color="green", label="plot figure")
plt.legend()

plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.show()
```

### 网格函数grid

```python
x = np.linspace(0.05, 10, 1000)
y = np.sin(x)

plt.plot(x, y, linestyle="-", color="green", linewidth=2, label="plot figure")
plt.legend(loc="upper right")

plt.grid(linestyle=":", color="red")

plt.show()
```

### 绘制水平线axhline

```python
x = np.linspace(0.05, 10, 100)
y = np.sin(x)

plt.plot(x, y, linestyle="-.", linewidth=2, color="c", label="plot figure")

plt.legend(loc="upper left")

plt.axhline(y=0.0, color="red", linestyle="--", lw=2)
plt.axvline(x=4.0, color="red", linestyle="--", lw=2)

plt.show()
```

### 绘制垂直于x轴的参考区域

```python
x = np.linspace(0.05, 10, 1000)
y = np.sin(x)

plt.plot(x, y, linestyle="-.", linewidth=2, color="c", label="plot figure")

plt.legend(loc="upper right")

plt.axvspan(xmin=4.0, xmax=6.0, facecolor="y", alpha=0.3)
plt.axhspan(ymin=0.0, ymax=0.5, facecolor="y", alpha=0.3)

plt.show()
```

### 添加图形内容细节的指向型注释文本annotate

```python
x = np.linspace(0.05, 10, 1000)
y = np.sin(x)

plt.plot(x, y, linestyle="-.", linewidth=2, color="c", label="plot figure")

plt.legend(loc="upper right")

plt.annotate(text="maximum",
             xy=(np.pi/2, 1.0),
             xytext=(np.pi/2+1.0, 0.8),
             weight="bold",
             color="red",
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3", color="blue"))

plt.show()
```

### 添加图形内容细节的无指向型注释文本text

```python
x = np.linspace(0.05, 10, 1000)
y = np.sin(x)

plt.plot(x, y, linestyle="-.", linewidth=2, color="c", label="plot figure")

plt.legend(loc="upper right")

plt.text(3.10, 0.09, "y=sin(x)", weight="bold", color="blue")

plt.show()
```

### 添加图形内容的标题title

```python
x = np.linspace(0.05, 10, 1000)
y = np.sin(x)

plt.plot(x, y, linestyle="-.", linewidth=2, color="c", label="plot figure")

plt.legend(loc="lower right")

plt.title("y=sin(x)")

plt.show()
```

### 添加图例legend

```python
x = np.linspace(0.05, 10, 1000)
y = np.sin(x)

plt.plot(x, y, linestyle="-.", linewidth=2, color="red", label="plot figure")

plt.legend(loc="lower left")

plt.show()
```
