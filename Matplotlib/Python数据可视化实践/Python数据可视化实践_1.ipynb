## Python数据可视化时间

```python
import numpy as np
import matplotlib.pyplot as plt
from jupyterthemes import jtplot
jtplot.style("gruvboxl")
```

### plot函数画趋势

```python
x = np.linspace(0.05, 10, 1000)
y = np.cos(x)

plt.plot(x, y, color="red", linestyle="-", linewidth=2, label="plot figure")
plt.legend()

plt.show()
```

### scatter函数

```python
x = np.linspace(0.05, 10, 1000)
y = np.random.rand(1000)

plt.scatter(x, y, label="scatter figure")
plt.legend(loc="upper left")

plt.show()
```

### xlim, ylim函数设置坐标轴范围

```python
x = np.linspace(0.5, 10, 1000)
y = np.random.rand(1000)

plt.scatter(x, y, label="scatter figure")
plt.legend(loc="upper right")

plt.xlim(0.05, 10)
plt.ylim(0, 1)

plt.show()
```

### xlabel函数设置

```python
x = np.linspace(0.05, 10, 1000)
y = np.sin(x)

plt.plot(x, y, linestyle="-", linewidth=1, color="green", label="plot figure")
plt.legend()

plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.show()
```

### 网格函数grid

```python
x = np.linspace(0.05, 10, 1000)
y = np.sin(x)

plt.plot(x, y, linestyle="-", color="green", linewidth=2, label="plot figure")
plt.legend(loc="upper right")

plt.grid(linestyle=":", color="red")

plt.show()
```
